```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("E:\\【ML_DL】\\【服装项目】\\0416\\上海地区-周销量\\data03.csv",encoding='utf-8',low_memory=False)
```

```python
df['销售日期'] = pd.to_datetime(df['销售日期'], format="%Y/%m/%d")
```

```python
df
```

```python
plt.plot(df['销售价格'],df['数量'],'b.')
plt.rcParams['font.sans-serif'] = ['SimHei'] 
# plt.title("标题")
plt.xlabel("销售价格")
plt.ylabel("数量")
```

```python
df['销售日期'] = range(1,279)
```

##### 2013年吊牌价格低中高与销售量

```python
from matplotlib import pyplot as plt
fig,ax=plt.subplots()
plt.rcParams['figure.figsize'] = (20.0,20.0)
ax.bar(df['销售日期'][:26],df['数量'][:26])
ax.set_xlabel('销售周数')  #设置x轴标签
ax.set_ylabel("每周的销售数量")  #设置y轴标签
ax.set_title("2013年每周的销售数量图")  #设置标题
plt.show()  #显示图像
```

```python
fig,ax=plt.subplots()
y1=df['吊牌价格_低'][:26]
y2=df['吊牌价格_中'][:26]
y3=df['吊牌价格_高'][:26]

ax.bar(df['销售日期'][:26],y1,color="g")
ax.bar(df['销售日期'][:26],y2,bottom=y1,color="r")
ax.bar(df['销售日期'][:26],y3,bottom=(y1+y2),color="blue")
plt.show()
```
