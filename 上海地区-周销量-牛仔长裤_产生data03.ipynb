```python
# name形如'address'，是字符串的形式
def encode_text_dummy(df, name, delete =  True):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = "{}-{}".format(name, x)
        df[dummy_name] = dummies[x]
    if delete:    
        df.drop(name, axis=1, inplace=True)
```

```python
encode_text_dummy(df, name = "颜色说明" )# 对货品名称进行独热编码
```

```python
encode_text_dummy(df, name = "尺码" )# 对货品名称进行独热编码
```

```python
encode_text_dummy(df, name = "吊牌价" )# 对货品名称进行独热编码
```
